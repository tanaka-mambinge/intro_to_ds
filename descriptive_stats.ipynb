```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.set_printoptions(suppress=True)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv("data/House_Rent_Dataset.csv")
```

```python
df.head()
```

```python
plt.hist(df["Rent"], bins=50)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Rent')
plt.show()
```

The histogram above is positively skewed

```python
log_t_rent = np.log(df["Rent"])

# Calculate the mean and standard deviation of the data
mu, std = stats.norm.fit(log_t_rent)

# Create a normal distribution with the same mean and standard deviation
dist = stats.norm(mu, std)

# Generate some x values to use for the normal distribution plot
x_vals = np.linspace(log_t_rent.min(), log_t_rent.max(), 100)

# Plot the normal distribution over the histogram
plt.plot(x_vals, dist.pdf(x_vals), 'r-', label='Normal Distribution')

# Add a legend and axis labels
plt.legend()
plt.hist(log_t_rent, bins=50, density=True, alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Rent')
plt.show()
```

After apply log transform, the data is now close to the normal distribution.
